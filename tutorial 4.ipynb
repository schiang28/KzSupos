```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
import scipy.optimize
import sklearn
```

```python
x = [7.2, 7.3, 7.8, 8.2, 8.8, 9.5]
n = len(x)
```

```python
def mu_hat(x):
    return np.mean(x)

sigma_hat = np.sqrt(np.mean((x-mu_hat(x))**2))
#mu_hat(x)

# simulate multi-universe

def resample():
    return np.random.normal(loc=mu_hat(x), scale=sigma_hat, size=n)

mu_ = [mu_hat(resample()) for _ in range(10000)]
mu_
```

```python
# confidence interval 95% of mu_hat
lo, hi = np.quantile(mu_, 0.025), np.quantile(mu_, 0.975)
lo,hi
```

```python
fig,ax = plt.subplots(figsize=(3,1.5))
ax.hist(mu_, bins=30, alpha=.4)
ax.axvline(lo, linestyle='dotted', color='0.4')
ax.axvline(hi, linestyle='dotted', color='0.4')
ax.set_yticks([])
plt.show()
```

```python
x = [4.3, 5.1, 6.1, 6.8, 7.4, 8.8, 9.9]
y = [8.3, 8.5, 8.9]
m = len(x)
n = len(y)
```

```python
# test statistics
def mu_hat(x):
    return np.mean(x)

def v_hat(y):
    return np.mean(y)

def difference(x,y):
    return v_hat(y) -mu_hat(x)

difference(x,y)
```

```python
sigma_hat = np.sqrt((np.sum((x-mu_hat(x))**2)+np.sum((y-v_hat(y))**2))/(m+n))
```

```python
sigma_hat
```

```python
# resample
def resample():
    return np.random.normal(loc=mu_hat(x), scale=sigma_hat, size=m), np.random.normal(loc=v_hat(y), scale=sigma_hat, size=n)

diff_ = [difference(*resample()) for _ in range(10000)]
#diff_

lo, hi = np.quantile(diff_, 0.025), np.quantile(diff_, 0.975)
lo,hi
```

```python
fig,ax = plt.subplots(figsize=(3,1.5))
ax.hist(diff_, bins=30, alpha=.4)
ax.axvline(lo, linestyle='dotted', color='0.4')
ax.axvline(hi, linestyle='dotted', color='0.4')
ax.set_yticks([])
plt.show()
```

```python
def mean_hat(x,y):
    return np.mean(np.concatenate([x,y]))

sigma_hat = np.sqrt(np.mean((np.concatenate([x,y]) - mean_hat(x,y))**2))
# 2.resample
def resample():
    return np.random.normal(loc=mean_hat(x,y), scale=sigma_hat, size=m),np.random.normal(loc=mean_hat(x,y), scale=sigma_hat, size=n)

diff_ = np.array([difference(*resample()) for _ in range(10000)])
p = 2 * np.min([np.mean(diff_ >= difference(x,y)), np.mean(diff_ <= difference(x,y))])
```

```python
p
```

```python
print(f"p={p:.3}")
t = diff_
fig,ax = plt.subplots(figsize=(3.5,1.5))
bins = np.linspace(-4.5, 4.5, 60)
ax.hist(t, bins=bins, alpha=.3)
ax.hist(t[t>=difference(x,y)], bins=bins, color='red')
ax.hist(t[t<=np.quantile(t,p/2)], bins=bins, color='red')
ax.axvline(x=difference(x,y), linestyle='dashed', color='black')
ax.set_yticks([])
plt.show()
```

```python
modis = pd.read_csv('https://www.cl.cam.ac.uk/teaching/2122/DataSci/data/modis.csv')
modis = modis.loc[modis.day_of_year==240].sort_values('year')
modis = modis.sort_values('year')

fig,ax = plt.subplots(figsize=(3.5,1.5))
ax.bar(modis.year, modis.num_fires, width=0.8, alpha=0.4, color='red')
plt.xticks(np.arange(2003,2020,2), rotation=-60)
plt.show()
```

```python
i = (modis.year >= 2011) & (modis.year <= 2019)
year,fires = modis.year.loc[i].values, modis.num_fires.loc[i].values
year, fires
```

```python
#find 95% confidence interval of delta
f =np.where(year==2019, 1,0)
feature_vectors = np.column_stack([f])
model = sklearn.linear_model.LinearRegression() # y-axis constant
model.fit(feature_vectors, fires)

def delta(x):
    model = sklearn.linear_model.LinearRegression() # y-axis constant
    model.fit(feature_vectors, x)
    (delta_hat,) = model.coef_
    return delta_hat

def resample():
    pred =  model.predict(feature_vectors)
    sigma_hat = np.sqrt(np.mean((fires-pred)**2))
    return np.random.normal(loc=pred, scale=sigma_hat)

delta_ =[delta(resample()) for _ in range(10000)]

#95% confidence interval
lo, hi = np.quantile(delta_, 0.025), np.quantile(delta_,0.975)

print(f"95% confint [{lo:.0f},{hi:.0f}]")

fig,ax = plt.subplots(figsize=(3.5,1.5))
ax.hist(delta_, bins=30, alpha=.4)
ax.axvline(lo, linestyle='dotted', color='0.4')
ax.axvline(hi, linestyle='dotted', color='0.4')
ax.set_yticks([])
plt.show()
```
